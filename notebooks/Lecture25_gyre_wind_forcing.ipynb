```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 2000
y = np.arange(0, 2000)

τ = -0.1*np.cos(np.pi*y/L)
```

```python
plt.figure(figsize=(4,5))
plt.plot(τ, y)
plt.ylim(0, L)
plt.axvline(0, linestyle="dashed", color="black", alpha=0.5)
plt.ylabel(r"meridional distance $y$ [km]")
plt.xlabel(r"zonal wind stress $\tau^{(x)}(y)$ [N/m$^{2}$]")
```
